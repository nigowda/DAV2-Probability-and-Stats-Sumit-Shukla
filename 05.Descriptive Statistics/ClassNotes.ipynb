**IQR To Detect Outliers**

```python
q1 = d['Runs'].quantile(0.25)
q3 = d['Runs'].quantile(0.75)
iqr = q3-q1
```

```python
l = q1-1.5*iqr
u = q3+1.5*iqr
d[(d['Runs']<l)| (d['Runs']>u)]
```

```python
q1 = s['Runs'].quantile(0.25)
q3 = s['Runs'].quantile(0.75)
iqr = q3-q1
```

```python
l = q1-1.5*iqr
u = q3+1.5*iqr
s[(s['Runs']<l)| (s['Runs']>u)]
```
